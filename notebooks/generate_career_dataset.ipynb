```python
rows = []

samples_per_career = 100

for career, skills in CAREERS.items():

    for _ in range(samples_per_career):

        n_skills = random.randint(
            4,
            min(7, len(skills))
        )

        selected_skills = random.sample(
            skills,
            n_skills
        )

        rows.append({
            "skills": " ".join(selected_skills),
            "career": career
        })

df = pd.DataFrame(rows)

print(df.shape)

df.head()
```

```python
df = df.sample(
    frac=1,
    random_state=42
).reset_index(drop=True)
```

```python
df.to_csv(
    "../data/career_dataset.csv",
    index=False
)

print(
    "Dataset saved successfully!"
)
```
